```python
import pandas as pd
import numpy as np
import csv
```

```python
#Limpiar valores vacios 
df = pd.read_csv('airports.csv',header=0,encoding='utf-8',na_filter=False)
df.head()
```

```python
df['elevation_ft'] = pd.to_numeric(df['elevation_ft'], errors='coerce')
df.loc[96].elevation_ft
```

```python
for column in df.columns:
    if df[column].dtype == object:  # Check if column contains text
        df[column] = df[column].fillna('').replace('', 'NULL')
    else:  # Column contains numeric values
        df[column] = df[column].fillna(0)

        
        
df.loc[96].elevation_ft
```

```python
df.head()
```

```python
df.to_csv('cleanAirports.csv', index=False)
```

```python
#agregar quotes
df = pd.read_csv('cleanAirports.csv',header=0,encoding='utf-8',na_filter=False)
df.to_csv('quotedAirports.csv', index=False, quoting=csv.QUOTE_NONNUMERIC, quotechar='"')
```
